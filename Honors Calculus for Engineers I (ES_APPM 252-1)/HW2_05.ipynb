HW2_05 - Bessel function

Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jv
```

Plotting Jâ‚„(x) from x = 0 to 50

```python
x = np.linspace(0, 50, 1000) #1000 points for smooth curve
y = jv(4, x) #J4 at each of the values of x
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Jâ‚„(x)')
plt.title('Bessel Function of the First Kind: Jâ‚„(x)')
plt.grid(True)
```

Finite Differences and Taylor series approximation

```python
x0 = 33
Dx = 0.01
J4_prime = (jv(4, x0 + Dx) - jv(4, x0 - Dx)) / (2 * Dx) #Finite difference derivative of J4 at x = 33
x = np.linspace(30, 36, 120)
J4_exact = jv(4, x)
J4_taylor = jv(4, x0) + J4_prime * (x - x0) #First-order Taylor series = J4(33) + J4'(33)*(x - 33)
plt.plot(x, J4_exact, label='Jâ‚„(x)') #plotting the functions
plt.plot(x, J4_taylor, label='Taylor approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Jâ‚„(x) and First-Order Taylor Approximation at x = 33')
plt.grid(True)
```

Newton's method

```python
error_bound = 1e-6 
x0 = 33.0
Dx = 0.01

while True:
    f_x = jv(4, x0) 
    derivative = (jv(4, x0 + Dx) - jv(4, x0 - Dx)) / (2 * Dx) #using finite difference to find derivative
    x_new = x0 - f_x / derivative #using Newton's method to find new x value
    
    if abs(x_new - x0) < error_bound: #end loop when change from new x to previous x is less than 1e-6
        x0 = x_new
        break
    
    x0 = x_new

print(f'Estimated zero: {x0}')
print(f'J4(estimated zero): {f_x}')
```

Estimated Zero: 33.5371377118. 

The iteration stops when the difference between the new x and the previous x differed by less than 1e-6. Since Newtonâ€™s method converges quadratically near the root, the remaining changes after this point are much smaller than 1e-6, so the total error is expected to be within 1e-6.

Additionally, the absolute value of the function evaluated at the estimated zero is 1.8263334460836605e-10 < 1e-6. This provides additional evidence that the estimated zero is within 1e-6 of the actual zero.
