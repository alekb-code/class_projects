HW1_05 - Fibonacci Polynomials

First, general imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

Function to evaluate the polynomial of nth degree at x

```python
def F_func(n, x):
    if n == 0:
        return 0 #given condition 1
    elif n == 1:
        return 1 #given condition 2
    else:
       return x*F_func(n-1, x) + F_func(n-2, x) #recurrence relation
```

Testing the function

```python
F_func(1, 6)
```

```python
F_func(30, 5)
```

```python
F_func(35, 10)
```

Making a plot of F10(x)

```python
n = 10 # can change to test other n values
x = np.linspace(-1, 1, num=100)
y = F_func(10, x)
plt.plot(x, y)
plt.title("Fibonacci Polynomial of 10th degree for x between -1 and 1")
plt.xlabel('X')
plt.ylabel(f'$F_{{{n}}}(X)$')
```

Verifying the first relation

```python
n = 10
m = 6
lhs_list=[] 
rhs_list=[]
for i in np.arange(-1, 1.1, 0.01): #testing all x values between -1 and 1
    lhs_list.append(round(F_func(m+n, i), 5)) #rounding to 5 decimal places to avoid decmial issues
    rhs_list.append(round((F_func(n+1, i)*F_func(m, i))+(F_func(n, i)*F_func(m-1, i)), 5))
print(lhs_list)
print(rhs_list)
is_true = lhs_list==rhs_list #boolean to test if the entries of corresponding x values are equal in each side of the equation
print(is_true)
```

Explaination: I tested 201 x values, from -1 to 1 ascending in increments of .01, into both the left and right hand sides of the given equation where n=10 and m=6. Every iteration of the loop, I added the outputs of each side into respective lists, in the same order. I made sure to round these values to a fixed amount of decmial places (5 in this case) to negate rounding and truncating effects of multiplying and adding floats together. At the end, I compared the lists in their entirety to each other, so that the boolean "is_true" would only be set to true if every one of the 201 entries matched. The boolean is set to true, meaning that __these expressions are equal in the desired range__ (when x is between -1 and 1, n is 10, and m is 6).

Testing the second relation

```python
import random
n = random.randint(0, 10) #random ints assigned to n and m
m = random.randint(0, 10)
lhs_list=[]
rhs_list=[]
for i in np.arange(-1, 1.1, 0.01):
    lhs_list.append(round(F_func(n+1, i)*(F_func(m, i)), 5))
    rhs_list.append(round((F_func(n, i))*(F_func(m+1, i)), 5))
print(n) #to see what ints were generated for n and m
print(m)
print(lhs_list) 
print(rhs_list)
is_true = lhs_list==rhs_list
print(is_true)

```

Explanation: In order to test this relation more thoroughly,  I added a random element to the n and m values, so that each time the script runs, it uses new values (between 1 and 10 for simplicity). I kept the same x range between -1 and 1 to use 201 points, and I went through the same exact process with comparing lists of outputs for each side of the equation. However, for this relation, we can see that "is_true" is false more often than true. Specifically, its false whenever n and m are set to different integers since when n=m, the equation is true algebraically. Because this relation is not always true for different values of n and m, __it does not hold__.
